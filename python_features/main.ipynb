#පයිතන් වල __ (double underscore) වලින් පටන් ගන්නා variables වලට විශේෂ අර්ථ තිබේ.

- __name__: Script එක direct run කරනවාද නැත්නම් module එකක් ලෙස import කරනවාද යන්න පෙන්වයි.

- __file__: දැනට ක්‍රියාත්මක වන script එකේ path එක පෙන්වයි.

- __all__: Module එකකින් export කරන දේවල් පාලනය කරයි.

# Understanding scope

```python
x = 1

def test():
    # global x
    x = 10
    print(x)

test()
print(x)

x = x + 5
print(x)
test()
```

පයිතන් variable එකක් හොයන පිළිවෙලක් තියෙනවා. ඒකට කියන්නේ LEGB Rule කියලා.

1. Local (Function එක ඇතුළේ)

2. Enclosing (Nested functions වල)

3. Global (Script එකේ ඉහළම මට්ටමේ)

4. Built-in (Python වල මුලින්ම එන keywords)

# The Assignment Expression (Walrus Operator :=)

```python
import pprint
x = 5
print(x)

(x := 10)
print(x)

while (thestr := input('Value? ')) != 'exit':
    print(thestr)

values = [12, 0, 10,5, 9, 18, 41, 23, 30, 16, 18, 9, 18, 22]
l =len(values)
s = sum(values)

val_data = {
    'length': l,
    'sum': s,
    'Average': s / l,
    'Max': max(values),
    'Min': min(values),
}

pprint.pp(val_data)
```

```python
import  pprint
from dataclasses import dataclass

values = ['one','two','three','four','five']

print(*values)

print(*values, sep=' -- ')

for i in range(0, len(values)):
    print(values[i], end=f'[line: {str(i+1)}]\n')
```
