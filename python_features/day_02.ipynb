#  Using Iterators
## Basic iterators

```python
days = ['Sun', 'Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat']
```

```python
for d in days:
    print(d)
```

```python
i = iter(days)
print(next(i))
print(next(i))
print(next(i))
```

```python
with open('testfile.txt', 'r') as fp:
    for line in iter(fp.readline, ''):
        print(line)
```

## More iterators

```python
for d in range(len(days)):
    print(d+1 ,days[d])
```

```python
for i,d in enumerate(days,start=1):
    print(i,d)
```

```python
daysFr = ['Dim', 'Lun', 'Mar', 'Mer', 'Jeu', 'Ven','Sum']
```

```python
for d in zip(days,daysFr):
    print(d)
```

```python
for i,d in enumerate(zip(days,daysFr),start=1):
    print(i,d[0],'=',d[1],'In French')
```

```python
import itertools
seq1 = ['A', 'B', 'C', 'D', 'E', 'F']
seq2 = [1,2,3,4]
seq3 = 'xyz'
```

```python
result = itertools.zip_longest(seq1, seq2, seq3, fillvalue='_')

print("Result :")
for item in result:
    print(item)

```



## For-else

```python
names = ['Harry', 'Ron', 'Messy', 'Jim', 'Kumara', 'Ben', 'Kevin']
```

```python
def find_name(target):
    for name in names:
        if name == target:
            print('Found a name!')
            return True
    else:
        print('Not found!')
        return False
```

```python
print(find_name('Harry'))
print(find_name('Ron'))
print(find_name('Rock'))
```

```python
def is_prime(num):
    for i in range(2,num):
        if num % i == 0:
            print(num,'Not prime!')
            break
    else:
        print(num,'is prime number!')
```

```python
is_prime(31)
is_prime(365)
```

## Itertools: count, cycle, accumulate

```python
names = ['Joe', 'Jane', 'Jim']
```

```python
cycler = itertools.cycle(names)
print(next(cycler))
print(next(cycler))
print(next(cycler))
print(next(cycler))
print(next(cycler))
```

```python
counter = itertools.count(100, 10)
print(next(counter))
print(next(counter))
print(next(counter))
print(next(counter))
print(next(counter))
print(next(counter))
```

```python
vals = [10, 20,30,40,50,40,30]
acc = itertools.accumulate(vals)
print(list(acc))
```

## Itertools: chain, chain.from_iterable, batched

```python
x = itertools.chain("ABCD", "1234")
print(list(x))
```

```python
s1 = 'ABCDEFG'
s2 = [1,2,3,4,5]
s3 = ['$', '%', '@', '&']

result = itertools.chain.from_iterable([s1,s2, s3])
print(list(result))
```

## Itertools: combinations and permutations

```python
cards = 'A23456789TJQK'
suits = 'SCHD'
deck = list(itertools.product(cards, suits))
```

```python
print(len(deck))
print(deck)
```

```python
teams = ('A','B','C','D')
```

```python
result = itertools.permutations(teams, 2)
print(list(result))
```

```python
result = itertools.combinations('ABCD', 3)
print(list(result))
```

```python
result = itertools.combinations_with_replacement('ABCD', 3)
print(list(result))
```
